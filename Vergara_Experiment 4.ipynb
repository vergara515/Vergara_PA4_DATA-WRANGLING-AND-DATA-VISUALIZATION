# ECE BOARD EXAM PROBLEM

In this problem, an ECE Board Exam Data data frame is loaded using Pandas. Data wrangling and Data visualization techniques will be used to analyze the data and present different (i) data frames; and (ii) visuals using the dataset given.

```python
#accessing the Pandas library
import pandas as pd
```

```python
#the variable df_ece is used to store the .csv file which contains the ECE Board Exam Data
df_ece = pd.read_csv('board2.csv')

#the df_ece variable is called to display the dataset from the .csv file
df_ece
```

# 1. Create the following data frames based on the format provided:

In this problem, the data set are filtered with multiple conditions to get the desired output needed. By the use of Pandas, the data are able to be filtered and cleaned. Sub-problems and questions to further demonsrate the process are shown below and explained.

# a. Instrumentation and Luzon Data Frame

Create a data frame, Instru, wherein 'Name', 'GEAS', and 'Electronics > 70' will be generated from the Instrumentation track and hometown Luzon.

1. Data in rows are filtered in the df_ece data frame.
2. After filtering, the code selected the specific columns from filtered rows.
3. The variable 'Instru' is used to contain the rows and columns that meets the filtering criteria.
4. A table wherein the 'Name's with their GEADS and a grade greater than 70 in Electronics is shown.

```python
#the data set in df_ece is filtered to what is needed through multiple conditions and stored to 'Instru'
Instru = df_ece[(df_ece['Track'] == 'Instrumentation')
        & (df_ece['Hometown'] == 'Luzon')
        & (df_ece['Electronics'] > 70)][['Name', 'GEAS', 'Electronics']]

#the Instru variable is called to display the data frame with filtered criteria
Instru
```

# b. Mindanao and Female Data Frame

Create a data frame, Mindy, wherein 'Name', 'GEAS', 'Electronics', and 'Average >= 55' will be generated from hometown Mindanao and gender Female.

1. Calculate the avearage score for each student given the grades for Math, Electronics, GEAS, and Communication.
2. Filtering the data set to get the Female students from Mindanao with an Average of equal or greater than 55.
3. The variable 'Mindy' is used to store the data frame with filtered conditions.
4. A table wherein Female students from Mindanao with an Average of equal or greater than 55 is shown.

```python
#calculating the average score for each student
df_ece['Average'] = df_ece[['Math', 'Electronics', 'GEAS', 'Communication']].mean(axis=1)

#the data set in df_ece is filtered to what is needed through multiple conditions is stored in variable 'Mindy'
Mindy = df_ece[(df_ece['Hometown'] == 'Mindanao')
        & (df_ece['Gender'] == 'Female')
        & (df_ece['Average'] >= 55)][['Name', 'Track', 'Electronics', 'Average']]

#the variable 'Mindy' is called to show the filtered data frame
Mindy
```

# 2. Create a visualization that shows how the different features contributes to average grade.

In this problem, graphs are created to visualize how different features contribute to average grade. Through the use of Seaborn and Matplotlib library, graphs are generated to further visualize the relationship (if there is) and effects of features to the average grade of each student.


```python
#accesing the Seaborn library
import seaborn as sns

#accesing the Matplotlib library
import matplotlib.pyplot as plt
```

```python
#setting figure size
plt.figure(figsize=(10, 6))

#creating box plot
sns.boxplot(x='Track', y='Average', hue='Gender', data=df_ece)

#set plot title
plt.title('Average Grade by Track and Gender')

#displaying the plot
plt.show()
```

```python
#setting figure size
plt.figure(figsize=(10, 6))

#
sns.boxplot(x='Hometown', y='Average', hue='Track', data=df_ece)
plt.title('Average Grade by Hometown and Track')

#
plt.show()
```

```python
#setting figure size
plt.figure(figsize=(10, 6))

#creating box plot
sns.boxplot(x='Hometown', y='Average', hue='Gender', data=df_ece)

#set plot title
plt.title('Average Grade by Hometown and Gender')

#displaying the plot
plt.show()
```

# Does chosen track in college, gender, or hometown contributes to a higher average score?

From the boxplots generated, it can be infered that the chosen track in college, gender, or hometown do contribute to a higher average score.

When it comes to the Instrumentation track, gender does not significantly impacts the average score; however, gender does have a noticeable impact in Communications and Microelectronics.

Based on the boxplot of 'Average Grade by Hometown and Track', it suggests that students from Luzon perform better on average, especially in Communication and Mictroelectronics tracks. When it comes to tracks, the Microelectronics track seems to yield higher average scores overall. In can also be observed that students with Microelectronics track and are from Luzon generally perform better on average.

Based on the boxplot of 'Average Grade by Hometown and Gender', it can be infered that the Female and Male students from Luzon have higher average scores than Visayas and Mindanao. On the other hand, Female students from Mindanao has the lowest range of average scores, while both Females and Males in Visayas perform averagely with one Female outlier.

In conclusion, by comparing all these boxplots altogether, it can be observed that the three factors- track, gender, and hometown- do contribute to the differences in average scores; however, their impact varies depending on the combination. But, it should also be noted that the only trait that consistently contributed to the higher average score, regardless of combination is Hometown. This is because Gender and Track did not matter when Hometown is involved as students in Luzon scored higher always. Moreover, the Microelectronics track performed better on average, while gender influenced the performance on the track.
